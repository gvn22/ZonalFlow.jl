# Validation of the GCE2 Method

Here we are validating the GCE2 method against statistics obtained from GQL. This seems straightforward enough, but a whole summer has been spent on addressing differences between QL and CE2, not to mention years spent on address differences between GQL and GCE2. Now we do seem to have a hold on the disagreements, so here's the method paper results

The barotropic equation for vorticity $\zeta = \nabla^2 \psi$ on the $\beta$-plane with driving and dissipation is:

\begin{equation}
        \partial_t \zeta = F(y,t) - \beta \frac{\partial \psi}{\partial x} + \nu \nabla^2 \zeta - \mu \zeta + \nu_\alpha \nabla^{2\alpha} \zeta - J(\psi,\zeta)
\end{equation}

where $F(y,t)$ is the driving term which is constant in the zonal ($x$) direction, $\beta(\theta)$ refers to the strength of transverse-variation of Coriolis force at latitude $\theta$, $\mu$ is the friction coefficient, $\nu$ is the kinematic viscosity, $\nu_\alpha$ is the hyperviscosity coefficient and $\alpha$ the corresponding exponent, $\psi$ is the streamfunction, and the Jacobian $J(A,B) \equiv \hat{z} \cdot \nabla A \times \nabla B = A_xB_y - A_yB_x$ where $\hat{z}$ is the unit vertical vector.

We are varying $F(y,t)$ over the cases below.

## Pointjet

Here the forcing function $F$ is of the form:

\begin{equation}
F = - ({\Xi}/{\tau}) \rm{tanh}[({\pi - y})/{\Delta y}]
\label{F1}
\end{equation}

where $\Xi$ is the strength and $\tau$ is the relaxation time. We choose $\Xi$ as a fraction of the planetary rotation $\Omega$, here we set $\Xi = \Omega$.

Parameters values are as follows

* $\Omega = 2\pi$
* $\mu = 0.05$
* $\nu = 0$
* $\nu_4 = 0$
* $\Xi = 1.0$
* $\Delta \theta = 0.1$
* $\tau$ = 20 days
